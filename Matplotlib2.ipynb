```python
#Color Maps:it is  a list of colors,where each has a value that range from 0-100 by using cmap with the value of the color map
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
v=[]
for i in range(0,100,7):
    v.append(i)
color=np.array([v])
plt.scatter(x,y, c=color, cmap='viridis')
plt.show()
```
