```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=([1,8,5])
y=([3,10,6])
plt.plot(x,y, 'g')
plt.title('Hour')
plt.xlabel('Month')
plt.ylabel('sin values')
```

```python
import numpy as np, math
```

```python
x=np.arange(0,math.pi*2,0.022)
y=np.sin(x)
plt.plot(x,y, 'r')
plt.title('S=sin graph')
plt.xlabel('angles')
plt.ylabel('sin values')
```

```python
year=[1991,1992,1993,1994,1995]
gdp=[300,400,600,900,700]
plt.plot(year,gdp, marker='o')
plt.title('GDP Graph for 5yrs')
plt.xlabel('years')
plt.ylabel('billions of RS')
```
