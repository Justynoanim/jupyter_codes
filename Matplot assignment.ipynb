```python
m=pd.read_csv("C:\\Users\\user\\Desktop\\Newdirtydata.csv")
m
```

```python
m.isnull().sum()
```

```python
plt.plot(m.Duration, m.Calories, 'o-r', ms=10,mec='k',mfc='b')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('A Graph of Duration vs Calories')
plt.show()
```

```python
plt.plot(m.Pulse, m.Calories, 'o-r', ms=10,mec='k',mfc='b')
plt.xlabel('Pulse')
plt.ylabel('Calories')
plt.title('A Graph of Pulse vs Calories')
plt.show()
```

```python
plt.plot(m.Maxpulse, m.Duration, 'o-r', ms=10,mec='k',mfc='b')
plt.xlabel('Duration')
plt.ylabel('Maxpulse')
plt.title('A Graph of Maxpulse vs Duration')
plt.show()
```
